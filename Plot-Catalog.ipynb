```python
data = ascii.read("Young-Objects-Compilation.csv", header_start=1, data_start=2)
```

```python
plt.scatter(data["Jmag"] - data["Kmag"], data["Jmag"]) # plot J-K vs. J
plt.ylim(reversed(plt.ylim())) # flip the y-axis
plt.xlabel("$J-K_s$", fontsize=20)
plt.ylabel("$J$", fontsize=20)
```

```python
ra = coord.Angle(data['RA'].filled(np.nan)*u.degree)
ra = ra.wrap_at(180*u.degree)
dec = coord.Angle(data['Dec'].filled(np.nan)*u.degree)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="mollweide")
ax.scatter(ra.radian, dec.radian)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="mollweide")
ax.scatter(ra.radian, dec.radian)
ax.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
ax.grid(True)
```

```python
fig.savefig("map.pdf")
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="mollweide")
ax.scatter(ra.radian, dec.radian)
ax.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
ax.grid(True)
sc = ax.scatter(ra.radian, dec.radian, c=data["Kmag"], cmap='viridis', s=20)
cbar = plt.colorbar(sc, orientation='vertical')
cbar.set_label('Kmag')
```

```python
# Aitoff projection
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="aitoff")
ax.scatter(ra.radian, dec.radian, c=data["Kmag"], cmap='viridis', s=20)
ax.set_xticklabels(['14h', '16h', '18h', '20h', '22h', '0h', '2h', '4h', '6h', '8h', '10h'])
ax.grid(True)
plt.show()
```

```python
# Hammer projection
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="hammer")
ax.scatter(ra.radian, dec.radian, c=data["Kmag"], cmap='viridis', s=20)
ax.set_xticklabels(['14h', '16h', '18h', '20h', '22h', '0h', '2h', '4h', '6h', '8h', '10h'])
ax.grid(True)
plt.show()
```

```python
# Lambert projection
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="lambert")
ax.scatter(ra.radian, dec.radian, c=data["Kmag"], cmap='viridis', s=20)
ax.set_xticklabels(['14h', '16h', '18h', '20h', '22h', '0h', '2h', '4h', '6h', '8h', '10h'])
ax.grid(True)
plt.show()
```

```python
# None (default) projection
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)
ax.scatter(ra.radian, dec.radian, c=data["Kmag"], cmap='viridis', s=20)
ax.set_xticklabels(['14h', '16h', '18h', '20h', '22h', '0h', '2h', '4h', '6h', '8h', '10h'])
ax.grid(True)
plt.show()
```
